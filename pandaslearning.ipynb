```python
import pandas  as pd
#series
data=[104,1230,150,170]
df=pd.Series(data, index=['a','b','c','d'])
print(df)
y=df.iloc[2]
print(y)
#this is used to change value from a series
df.loc['a'] =500
print(df)
#this will print values which are greater that 200 in the series
print(df[df>200])
#other method of making a series is
cows={'ayrshire':30 ,'jersey':40, 'Freshian':50 ,'Bulls':60}
my_cows=pd.Series(cows)
#print(my_cows)
my_cows.loc['Freshian'] +=20
#print(my_cows)
my_cows['calfs']=30
print(my_cows)

```

```python
#DATAFrAMES this are 2-D tabular information
golf_scores= {'names':['Brain Kiprono','Gilbert Kungu','Peter Kimani','dennis Kipragat'], 'Golf_course_played':['Thika golf club','Thika greens','Karen golf course','Nyeri golf course'],'Number_of_swings':[15,17,18,20]}
Golf_results=pd.DataFrame(golf_scores,index=[1,2,3,4]) 
#this is how to we make a mew Row
T=pd.DataFrame([{'names':'John Maina','Golf_course_played':'Ruiru spot club','Number_of_swings':30,'Medal_awarded':'GOLD'}],index=[5]) 
y=pd.concat([Golf_results,T])
print(y)
#adding a new column
Golf_results['Medal_awarded']=['Gold','Bronze','silver','Daimond']
print(pd.concat([Golf_results,T]))
```
